```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("hubble_data.csv")
data.head()
```

```python
data.columns=["distance","velocity"]
data.head()
```

```python
data.describe()
```

```python
dist=data['distance'].values.reshape(-1,1)
velocity=data['velocity'].values
```

```python
reg=LinearRegression().fit(dist,velocity)
```

```python
pred=reg.predict(dist)
```

```python
plt.scatter(dist,velocity,color="blue",label="données")
plt.plot(dist,pred, color="black",label="Prédiction")
plt.title("Regression linéare",color="blue")
plt.legend(loc="lower right")
plt.xlabel("distance")
plt.ylabel("vélocité")
plt.show()
```
